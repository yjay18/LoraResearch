# Day 19 Visualization Notebook

This notebook regenerates the Day 19 figures from the saved Day 13-18 artifacts.


```python
from pathlib import Path
from lorasurgeon.viz import (
    ADAPTERS,
    load_json,
    plot_feature_domain_heatmap,
    plot_top_features_bar,
    plot_universal_features_heatmap,
    plot_volcano_style,
)

label_root = Path('results/labels')
classification_root = Path('results/classification')
figure_root = Path('results/figures')
figure_root.mkdir(parents=True, exist_ok=True)
```

```python
for adapter in ADAPTERS:
    payload = load_json(label_root / f'{adapter}_feature_labels.json')
    plot_feature_domain_heatmap(payload, figure_root / f'{adapter}_domain_heatmap.png')
    plot_top_features_bar(payload, figure_root / f'{adapter}_top20_bar.png')
    plot_volcano_style(
        classification_root / f'{adapter}_classified_features.csv',
        classification_root / f'{adapter}_classification_summary.json',
        figure_root / f'{adapter}_volcano.png',
        adapter,
    )
```

```python
universal_summary = load_json(Path('results/day18_universal_features.json'))
plot_universal_features_heatmap(
    universal_summary,
    classification_root,
    figure_root / 'universal_features_heatmap.png',
)
```
